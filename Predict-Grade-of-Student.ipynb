# Chuẩn hóa dữ liệu
Đoạn mã này sử dụng StandardScaler từ thư viện scikit-learn để chuẩn hóa dữ liệu. Việc chuẩn hóa là một bước rất quan trọng trong quy trình học máy, đặc biệt là đối với các mô hình nhạy cảm với sự khác biệt về đơn vị đo của các đặc trưng (như SVM, KNN, Logistic Regression,...).

StandardScaler: Đây là một công cụ trong scikit-learn dùng để chuẩn hóa các đặc trưng của dữ liệu bằng cách tạo ra dữ liệu có phân phối chuẩn (normal distribution) với:
- Mean (trung bình) = 0.
- Standard Deviation (độ lệch chuẩn) = 1.

**Công thức chuẩn hóa**

Xét lần lượt từng cột thuộc tính, với mỗi x thì ta chuẩn hóa thành $x_{\text{scaled}}$ như sau:
$$
x_{\text{scaled}} = \frac{x - \mu}{\sigma}
$$

Trong đó:
- 𝑥 là giá trị của dữ liệu.
- N là số mẫu dữ liệu, ở đây cụ thể bằng 569
- $\mu$ là giá trị trung bình của đặc trưng, được tính như sau:
$$
\mu = \frac{x_1 + x_2 + x_3 + \dots + x_N}{N} 
$$

- $\sigma$ là độ lệch chuẩn của đặc trưng, được tính như sau:
$$
\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2}
$$

```python
from sklearn.preprocessing import StandardScaler

s = StandardScaler()
X_train = s.fit_transform(X_train)
X_test = s.fit_transform(X_test)


```

# Đánh giá mô hình hồi quy


Đây là 3 phương pháp phổ biến đánh giá mô hình hồi quy:

**Mean Absolute Error** (MAE) là trung bình trị tuyệt đối lỗi:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) là trung bình bình phương lỗi:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) là căn trung bình bình phương lỗi:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

Comparing these metrics:

- **MAE** dễ hiểu nhất vì nó là trung bình lỗi.
- **MSE** phổ biến hơn MAE, vì MSE "phạt" các lỗi lớn hơn, điều này thường hữu ích trong thế giới thực.
- **RMSE** phổ biến hơn MSE, vì RMSE có thể diễn giải được trong đơn vị của "y" .

Mục tiêu là cần cực tiểu các hàm lỗi trên

## Đánh giá mô hình hồi quy tuyến tính



Đây là 3 phương pháp phổ biến đánh giá mô hình hồi quy:

**Mean Absolute Error** (MAE) là trung bình trị tuyệt đối lỗi:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) là trung bình bình phương lỗi:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) là căn trung bình bình phương lỗi:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

So sánh giữa các ma trận

- **MAE** dễ hiểu nhất vì nó là trung bình lỗi.
- **MSE** phổ biến hơn MAE, vì MSE "phạt" các lỗi lớn hơn, điều này thường hữu ích trong thế giới thực.
- **RMSE** phổ biến hơn MSE, vì RMSE có thể diễn giải được trong đơn vị của "y" .

Mục tiêu là cần cực tiểu các hàm lỗi trên

```python
from sklearn import metrics 
print("MAE",metrics.mean_absolute_error(y_test,predictions))
print("MSE",metrics.mean_squared_error(y_test,predictions))
print("RMSE",metrics.root_mean_squared_error(y_test,predictions))

```

## Hồi quy Ridge

Hồi quy Ridge thêm một thuật ngữ phạt (penalty) vào hàm mất mát để làm giảm sự phức tạp của
mô hình, giúp các hệ số hồi quy ổn định hơn:

$$J_{ridge}(\omega) = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2 + \alpha \sum_{j=1}^{p}\omega_j^2$$

*   $\alpha\sum_{j=1}^{p}\omega_j^2$:
    *   Đây là thuật ngữ điều chuẩn L2 (L2 regularization).
    *   $\alpha$: Tham số điều chỉnh (regularization parameter).
    *   $\alpha$ lớn $\rightarrow$ tăng phạt vào $\omega_j^2$ $\rightarrow$ các hệ số nhỏ lại.
    *   $\alpha$ nhỏ $\rightarrow$ gần giống hồi quy tuyến tính.
*   Mục tiêu: Tìm điểm cân bằng giữa:
    1.  Tối ưu hóa sai số dự đoán.
    2.  Giảm độ lớn của các hệ số $\omega$, tránh overfitting.

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import RandomizedSearchCV
```

```python
ridge = Ridge() 
```

param_dist : Tham số cần tìm kiếm
-   alpha : lấy giá trị từ 0.001 đến 1000 theo thang Logarit
-   fit_intercept : Tùy chọn lấy hệ số chặn
-   solver : Thuật toán giải phương trình
-   'auto' : sẽ tự động lựa chọn thuật toán cho ra kết quả tốt nhất
-   'svd': Dùng Phân tích giá trị riêng (Singular Value Decomposition).
-   'cholesky': Dùng phương pháp Cholesky.
-   'saga': Là một thuật toán gradient descent tối ưu, hiệu quả cho dữ liệu lớn hoặc dạng sparse (thưa).

```python
param_dist = {
    'alpha': np.logspace(-3,3,100),
    'fit_intercept' : [True,False],
    'solver': ['auto', 'svd', 'cholesky', 'saga']
}
```

-   ridge: Mô hình Ridge Regression.
-   param_dist: Tham số cần tối ưu.
-   n_iter: Số tổ hợp tham số sẽ thử.
-   cv=5: Sử dụng kiểm tra chéo 5 lần để đánh giá mô hình.
-   scoring: Sử dụng MSE để đánh giá, nhưng đổi dấu để phù hợp với thuật toán tối ưu.
-   random_state: Đảm bảo tính tái lặp của kết quả.

```python
random_search = RandomizedSearchCV(
    ridge,
    param_distributions=param_dist,
    n_iter=50,
    cv = 5,
    scoring='neg_mean_squared_error',
    random_state=101  
)
random_search.fit(X_train, y_train)
```

## Hồi quy Lasso

**Công thức Hồi quy Lasso**

Hồi quy Lasso tối thiểu hóa hàm mục tiêu sau:

$$J(\omega) = \frac{1}{2N}\sum_{i=1}^{N}(y_i - \hat{y}_i)^2 + \lambda\sum_{j=1}^{p}|\omega_j|$$

*   $\frac{1}{2N}\sum_{i=1}^{N}(y_i - \hat{y}_i)^2$: Sai số bình phương trung bình (Mean Squared Error - MSE), hay còn gọi là lỗi dự đoán.
*   $\lambda\sum_{j=1}^{p}|\omega_j|$: Thuật ngữ regularization L1 (L1 penalty), phạt theo tổng độ lớn tuyệt đối của các hệ số.
*   $\lambda$: Hệ số điều chỉnh (regularization strength):
    *   Giá trị $\lambda$ càng lớn $\rightarrow$ regularization càng mạnh $\rightarrow$ nhiều hệ số bị làm bằng 0.
    *   Giá trị $\lambda = 0$ $\rightarrow$ Lasso trở thành hồi quy tuyến tính thông thường (không có ràng buộc).

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()
```

param_grid: chọn tham số cho mô hình hồi quy Lasso 
-   'alpha': np.logspace(-2,2,1000) tham số alpha sẽ được duyệt từ 0,01 đến 100 với 1000 giá trị
grid_search: sử dụng phương pháp grid_search để tối ưu hóa tham số
-   lasso: mô hình lựa chọn là hồi quy Lasso
-   param_grid: tham số truyền vfo
-   cv = 5: sử dụng kiểm tra chéo 5 lần
-   scoring='neg_mean_squared_error': tối ưu hàm âm MSE  


```python
param_grid = {
    'alpha' : np.logspace(-2,2,1000)
}
grid_search =GridSearchCV(
    lasso,
    param_grid,
    cv = 5,
    scoring='neg_mean_squared_error'
)
grid_search.fit(X_train,y_train)
best_lasso = grid_search.best_estimator_
```

 **Lý thuyết Random Forest Regressor**

Random Forest Regressor (RFR) là một mô hình hồi quy sử dụng tập hợp nhiều cây quyết định (Decision Trees) để dự đoán giá trị đầu ra. Đây là một phương pháp thuộc nhóm ensemble learning giúp cải thiện độ chính xác và giảm overfitting so với một cây quyết định đơn lẻ.

*   **1. Nguyên lý hoạt động**

    RFR hoạt động theo 3 bước chính:

     **Bước 1: Bootstrap Sampling (Lấy mẫu ngẫu nhiên có hoàn lại)**
    *   Mô hình sẽ tạo ra n cây quyết định (Decision Trees).
    *   Mỗi cây được huấn luyện trên một tập con ngẫu nhiên của dữ liệu gốc (Bootstrap Sampling).
    *   Mỗi tập con này chỉ chứa khoảng 63% dữ liệu ban đầu, còn lại sẽ được dùng để kiểm tra mô hình (Out Of Bag Error - OOB).

     **Bước 2: Xây dựng từng cây quyết định**
    *   Với mỗi cây, tại mỗi nút, RFR sẽ chọn một tập con nhỏ các đặc trưng (không phải toàn bộ đặc trưng).
    *   Chia nhánh dựa trên tiêu chí giảm phương sai (variance reduction) thay vì giảm độ bất thuần như trong phân loại.
    *   Một số tiêu chí thường dùng để chia nút:
        *   Mean Squared Error (MSE): $MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$
        *   Mean Absolute Error (MAE)
        *   Poisson Loss (cho dữ liệu phân phối Poisson)

     **Bước 3: Tổng hợp dự đoán từ nhiều cây**
    *   Mỗi cây sẽ đưa ra một giá trị dự đoán riêng.
    *   Kết quả cuối cùng được tính bằng trung bình cộng của tất cả dự đoán:
    
    $$\hat{y} = \frac{1}{n}\sum_{i=1}^n y_i$$
